```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sales = pd.read_csv('Datasets/gameSales2016.csv', sep=',')
```

```python
sales.head()
```

## Platform category

```python
sales.groupby(by="Platform").mean()
```

<font color='blue'><b><i>It's a bit hard to make a quick judgement when there is so much numerical data like this, but it seems like there are significant differences in most columns when it comes to platform

```python
barplot = sns.barplot(x="Platform", y="Global_Sales", data=sales)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90);
```

<font color='blue'><b><i>It looks like the NES and GB platforms have by far the most sales out of the rest. Most platforms sell roughly the same amount, with a few having more or less sales. There are definitely statistically significant differences between a number of platforms

```python
barplot = sns.barplot(x="Platform", y="Critic_Score", data=sales)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90)
barplot.set(ylim=(60, 95));
```

<font color='blue'><b><i>It seems like there are significant stastical differences between a couple of genres when it comes to critic score, except for the DC which seems to have a very great difference

## Genre category

```python
sales.groupby(by="Genre").mean()
```

<font color='blue'><b><i>It seems like there are major differences in a couple of columns when it comes to genre

```python
barplot = sns.barplot(x="Genre", y="Global_Sales", data=sales)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90);
```

<font color='blue'><b><i>It seems like there are significant stastical differences between a couple of genres when it comes to global sales. Platformers and shooters differ the most

```python
barplot = sns.barplot(x="Genre", y="Critic_Score", data=sales)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90)
barplot.set(ylim=(60, 80));
```

<font color='blue'><b><i>It seems like there are significant stastical differences between a couple of genres when it comes to critic score. The sports, role-playing and strategy games seem to score better than games of other genres
