```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
sales = pd.read_csv('Datasets/gameSales2016.csv', sep=',')
```

```python
sales.head()
```

## Platform VS Genre

<font color='blue'><b><i>I expect there to be barely any correlation because every platform has all kinds of genres of games

```python
contingencyTable = sales.groupby(['Platform','Genre']).size().unstack('Platform', fill_value=0)
contingencyTable
```

```python
contingencyTable.plot(kind='bar')
```

```python
chi2_contingency(contingencyTable)
```

<font color='blue'><b><i>It seems like there is no correlation at all because the chi-squared is 0
